```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('Baltimore911.csv')

# Data Cleaning
# Handle missing values by filling them with appropriate values (e.g., mean, median, or a specific value)
df.fillna(df.mean(), inplace=True)  # Replace missing values with column mean

# Detect and handle outliers (you can adjust the outlier detection and treatment method as needed)
def remove_outliers(df, columns, threshold=3):
    for column in columns:
        z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
        df = df[z_scores < threshold]
    return df

# Specify columns to check for outliers (numeric columns)
numeric_columns = df.select_dtypes(include=['number']).columns
df = remove_outliers(df, numeric_columns)

# Handle the problematic column with mixed data types
# In this example, we convert all values to strings
problematic_column = 'column_name'  # Replace 'column_name' with the actual column name causing the issue
df[problematic_column] = df[problematic_column].astype(str)

# Now, you can proceed with the dataset profiling as previously described
num_rows, num_columns = df.shape

profile_df = pd.DataFrame(columns=['Column Name', 'Data Type', 'Min', 'Max', 'Mean', 'Median', 'Std Dev', 'Unique Values', 'Top 3 Values', 'Missing %'])

for column in df.columns:
    data_type = df[column].dtype
    unique_values = df[column].nunique()
    missing_percent = (df[column].isnull().sum() / num_rows) * 100

    if data_type in ['int64', 'float64']:  # Numeric attributes
        min_val = df[column].min()
        max_val = df[column].max()
        mean_val = df[column].mean()
        median_val = df[column].median()
        std_dev = df[column].std()
        top_3_values = df[column].value_counts().head(3).index.tolist()

    elif data_type == 'object':  # Categorical attributes
        min_val = None
        max_val = None
        mean_val = None
        median_val = None
        std_dev = None
        top_3_values = df[column].value_counts().head(3).index.tolist()

    else:  # Other data types (temporal, spatial, etc.)
        min_val = None
        max_val = None
        mean_val = None
        median_val = None
        std_dev = None
        top_3_values = None

    profile_df = profile_df.append({
        'Column Name': column,
        'Data Type': data_type,
        'Min': min_val,
        'Max': max_val,
        'Mean': mean_val,
        'Median': median_val,
        'Std Dev': std_dev,
        'Unique Values': unique_values,
        'Top 3 Values': top_3_values,
        'Missing %': missing_percent
    }, ignore_index=True)

# Display the dataset profile table
profile_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'Baltimore911.csv' with the actual path)
df = pd.read_csv('Baltimore911.csv')

# Assuming you have a 'CrimeDate' column containing the date of each call
# You might need to convert it to a datetime object if it's not already in that format
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'])

# Extract year, month, and day-of-week information from the 'CrimeDate' column
df['Year'] = df['CrimeDate'].dt.year
df['Month'] = df['CrimeDate'].dt.month
df['DayOfWeek'] = df['CrimeDate'].dt.dayofweek  # Monday=0, Sunday=6

# Group the data by year, month, and day-of-week and count the incidents
yearly_incidents = df.groupby('Year').size()
monthly_incidents = df.groupby('Month').size()
day_of_week_incidents = df.groupby('DayOfWeek').size()

# Create subplots for year-to-year, monthly, and day-of-week plots
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

# Year-to-Year Plot
axes[0].plot(yearly_incidents.index, yearly_incidents.values, marker='o')
axes[0].set_title('Year-to-Year Crime Incidents')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Number of Incidents')

# Monthly Plot
axes[1].plot(monthly_incidents.index, monthly_incidents.values, marker='o')
axes[1].set_title('Monthly Crime Incidents')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Number of Incidents')
axes[1].set_xticks(range(1, 13))
axes[1].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Day-of-Week Plot
day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
axes[2].plot(day_of_week_incidents.index, day_of_week_incidents.values, marker='o')
axes[2].set_title('Day-of-Week Crime Incidents')
axes[2].set_xlabel('Day of Week')
axes[2].set_ylabel('Number of Incidents')
axes[2].set_xticks(range(7))
axes[2].set_xticklabels(day_names)

plt.tight_layout()
plt.show()
```

```python
#question 4
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'Baltimore911.csv' with the actual path)
# Specify data types for columns with mixed data types if known
dtype_mapping = {"column_name7": str}  # Replace "column_name7" with the actual column name

# Read the CSV file, specifying data types for problematic columns and setting low_memory=False
df = pd.read_csv('Baltimore911.csv', dtype=dtype_mapping, low_memory=False)

# Assuming you have a 'CrimeDate' column containing the date of each call
# You might need to convert it to a datetime object if it's not already in that format
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'])

# Extract year, month, and day-of-week information from the 'CrimeDate' column
df['Year'] = df['CrimeDate'].dt.year
df['Month'] = df['CrimeDate'].dt.month
df['DayOfWeek'] = df['CrimeDate'].dt.dayofweek  # Monday=0, Sunday=6

# Group the data by year, month, and day-of-week and count the incidents for each crime type
yearly_crime_type_counts = df.groupby(['Year', 'Description']).size().unstack(fill_value=0)
monthly_crime_type_counts = df.groupby(['Month', 'Description']).size().unstack(fill_value=0)
day_of_week_crime_type_counts = df.groupby(['DayOfWeek', 'Description']).size().unstack(fill_value=0)

# Plot Year-to-Year Variation
yearly_crime_type_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Year-to-Year Crime Type Distribution (All Locations)')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Plot Month-to-Month Variation
monthly_crime_type_counts.plot(kind='line', figsize=(12, 6))
plt.title('Monthly Crime Type Distribution (All Locations)')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()

# Plot Day-of-Week Variation
day_of_week_crime_type_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Day-of-Week Crime Type Distribution (All Locations)')
plt.xlabel('Day of Week')
plt.ylabel('Number of Incidents')
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], rotation=0)
plt.tight_layout()
plt.show()
```

```python
#question 5
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px

# Load your dataset (replace 'Baltimore911.csv' with the actual path)
df = pd.read_csv('Baltimore911.csv')

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the app layout
app.layout = html.Div([
    html.H1("Crime Data Exploration Dashboard"),
    
    # Dropdown for selecting location
    html.Label("Select Location:"),
    dcc.Dropdown(
        id='location-dropdown',
        options=[{'label': location, 'value': location} for location in df['Location'].unique()],
        multi=True
    ),
    
    # Slider for selecting time period (year)
    html.Label("Select Year:"),
    dcc.RangeSlider(
        id='year-slider',
        min=df['Year'].min(),
        max=df['Year'].max(),
        step=1,
        marks={str(year): str(year) for year in range(df['Year'].min(), df['Year'].max() + 1)},
        value=[df['Year'].min(), df['Year'].max()]
    ),
    
    # Scatter map for spatial exploration
    dcc.Graph(id='spatial-map'),
    
    # Bar chart for temporal exploration
    dcc.Graph(id='temporal-chart')
])

# Define callback to update spatial map
@app.callback(
    dash.dependencies.Output('spatial-map', 'figure'),
    [dash.dependencies.Input('location-dropdown', 'value')]
)
def update_spatial_map(selected_locations):
    filtered_df = df[df['Location'].isin(selected_locations)]
    fig = px.scatter_mapbox(
        filtered_df,
        lat='Latitude',
        lon='Longitude',
        hover_name='Description',
        hover_data=['CrimeDate'],
        color='Description',
        title="Spatial Pattern of Crime Incidents",
        mapbox_style="open-street-map",
        zoom=10
    )
    return fig

# Define callback to update temporal chart
@app.callback(
    dash.dependencies.Output('temporal-chart', 'figure'),
    [dash.dependencies.Input('year-slider', 'value')]
)
def update_temporal_chart(selected_years):
    filtered_df = df[(df['Year'] >= selected_years[0]) & (df['Year'] <= selected_years[1])]
    fig = px.bar(
        filtered_df.groupby('Month').size().reset_index(),
        x='Month',
        y=0,
        labels={'0': 'Number of Incidents'},
        title="Monthly Variation of Crime Incidents"
    )
    fig.update_xaxes(title="Month", tickvals=list(range(1, 13)), ticktext=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
